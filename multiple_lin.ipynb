```python
df = pd.read_csv("train.csv")
```

```python
def clean_data(df):
    df_num_cl = df.select_dtypes(include='number').columns.to_list()
    df[df_num_cl].fillna(0,inplace=True)
    df_obj_cl = df.select_dtypes(exclude='number').columns.to_list()
    df[df_obj_cl].fillna('other',inplace=True)
    return df
```

```python
def normalize_data(df):
    num_cols = df.select_dtypes(include='number').columns
    scaler = MinMaxScaler()
    
    # Returns a new dataframe with scaled columns
    return df.assign(**{col: scaler.fit_transform(df[[col]]).flatten() for col in num_cols})
```

```python
def encoding(df):
    df_obj_col = df.select_dtypes(include='object').columns.to_list()
    le = LabelEncoder()
    df[df_obj_col] = df[df_obj_col].apply(le.fit_transform)
    return df
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer
```

```python
data_cleaning_pipeline = Pipeline([
    ('base_data_cleaning',FunctionTransformer(clean_data)),
    ('normalization',FunctionTransformer(normalize_data)),
    ('encoding',FunctionTransformer(encoding))
])
```

```python
new_df = data_cleaning_pipeline.fit_transform(df)
```
